```python
%matplotlib notebook
import subprocess
import os
import sys
import matplotlib.pyplot as plt
import time
from mpl_toolkits.mplot3d import Axes3D
```

```python
exe = "PFT.x"
d = '/home/jobasha/Documents/Dropbox/Repositories/planningfortransparency/problems/grid-obst/'
output_dir = 'results/'
output_file = 'results.txt'
```

```python
object_vector_l = []
time_first_action_vector_l = []
domain = d + "domain.pddl"
threshold_1_v = []
threshold_2_v = []
threshold_3_v = []
count = 0
dirs = list(sorted(os.listdir(d)))

ns = []
hs= []
ths = []
tis = []

for problem_n in dirs:
     #print(problem_n)
    try:

#         if(not problem_n.startswith("prob")):
#             print("\t\tReject: " + problem_n)
#             continue
#         count = 1 + count
#         problem_file = d + problem_n
#         numbers = problem_n.split('_')

        if(not problem_n.startswith("problem")):
            print("\t\tReject: " + problem_n)
            continue
        count = 1 + count
        problem_file = d + problem_n
        numbers = problem_n.split('_')

        n = int(numbers[1])
        m = int(numbers[2])
        #prob_number = int(numbers[4])
        hype_number = int(numbers[3])
        prob_number = int(numbers[4].split(".")[0])
        objects = n*m
        print(str(count) + ". Problem - " + str(n) + "x" +str(m) + "-"+ str(prob_number))
        sys.stdout.flush()

        hype = d + "hyp_" + str(n) + "_" +str(m) + "_" + str(hype_number) + ".dat"

    #     print("\t" +problem_file)

        problem = problem_file

        cmd = ["time", "python"]
        flags = [ "/home/jobasha/Documents/PhD/Planners/FastDownward/fast-downward.py","--alias", "seq-sat-lama-2011", domain, problem]

        pipe = subprocess.Popen(
            cmd + flags,
            stdin=subprocess.PIPE, stdout=subprocess.PIPE, stderr=subprocess.PIPE)
        #print(''.join(pipe.stderr.readlines()))
        answer, stderr = pipe.communicate()
        #print(answer)
        #print(stderr)
        answer1 = stderr.decode().split("\n")
        answer2 = answer1[2].split("system")
        answer3 = answer2[0].split("user")
        time_taken = sum(list(map(float, answer3 )))


        file = open("sas_plan.1","r")
        counter = 0
        plan = []
        for lines in file :
            counter = counter + 1
            plan.append(lines)

        cmdPR = ["/home/jobasha/Documents/Dropbox/Repositories/planningfortransparency/utilities/plan_recognition/./PRM.x"]
        flagsPR = [domain, problem, hype, "sas_plan.1", "-b", "10000000", "-a"]
        #print(domain, problem, hype, "sas_plan.1")
        pipePR = subprocess.Popen(
            cmdPR + flagsPR,
            stdin=subprocess.PIPE, stdout=subprocess.PIPE, stderr=subprocess.PIPE)
        #print(''.join(pipe.stderr.readlines()))
        answerPR, stderrPR = pipePR.communicate()

        #print(answerPR.decode())
        #print(stderrPR.decode())

        lines = str(answerPR).split("\\n")[1:-1]
        #print(lines)
        threshold_1_v.append(float(lines[0]))
        threshold_2_v.append(float(lines[1]))
        threshold_3_v.append(float(lines[2]))


        print("Problem - " + str(n) + "-" + str(prob_number) + "-" + ", Cost: " + str(counter -1) + ", " + str(time_taken) + ", " + lines[0] )
        #print(str(counter-1))
        dst = "../lama/" + problem_n + ".res"
        os.rename("sas_plan.1", dst)
        os.remove("output")
        os.remove("output.sas")
        
        ns.append(n)
        hs.append(prob_number)
        ths.append(lines[0])
        tis.append(time_taken)
        
#         if(int(m) == 10 and int(h) == 1 and int(pth) == 3):
#             for pf in plan:
#                 print(pf)
        
#         if(float(lines[0]) > 1):
#             for pf in plan:
#                 print(pf)
        

        if(counter != 1):
            time_first_action_vector_l.append(time_taken)
            object_vector_l.append(objects)
    except BaseException as e:
        #print(answer.decode())
        print("Python Error: ", e)


        
    
print(sum(threshold_1_v) / len(threshold_1_v))
print(sum(threshold_2_v) / len(threshold_2_v))
print(sum(threshold_3_v) / len(threshold_3_v))

print()
for i in ns:
    print(i)
print()
for i in hs:
    print(i)
print()
for i in ths:
    print(i)
print()
for i in tis:
    print(i)
print()

#     except:


#         continue
```
